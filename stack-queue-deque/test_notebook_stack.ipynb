```python
class Empty(Exception):
    pass
class Stack:
    def push(self, item):
        pass
    def pop(self):
        pass
    def top(self):
        pass
    def is_empty(self):
        return True
    def size(self):
        pass
    def __len__(self):
        pass
    def __check_is_empty(self):
        if self.is_empty():
            raise Empty("Stack is empty")
class ArrayStack(Stack):
    def __init__(self):
        self.__array = []
        self.__total_ops = 0
    def __total_ops_pp(self):
        self.__total_ops+=1
    def __check_is_empty(self):
        if self.is_empty():
            raise Empty("Stack is empty")
    def push(self, item):
        self.__total_ops_pp()
        self.__array.append(item)
    def pop(self):
        self.__check_is_empty()
        self.__total_ops_pp()
        return self.__array.pop()
    def top(self):
        self.__check_is_empty()
        self.__total_ops_pp()
        return self.__array[-1]
    def is_empty(self):
        self.__total_ops_pp()
        return len(self.__array) == 0
    def size(self):
        self.__total_ops_pp()
        return len(self.__array)
    def __len__(self):
        return len(self.__array)
    def __str__(self):
        return '# Contents: ' + str(self.__array) + '\n# Total Operations: ' + str(self.__total_ops) 
    def __iter__(self):
        return iter(self.__array)
    def __next__(self):
        return next(self.__array)
    def __str__(self):
        return '# Contents: ' + str(self.__array) + '\n# Total Operations: ' + str(self.__total_ops)
    def __repr__(self):
        return '# Contents: ' + str(self.__array) + '\n# Total Operations: ' + str(self.__total_ops)

def test_array_stack():
    s = ArrayStack()
    s.push(1)
    print(s.top())
    print(len(s))
    print(s.size())
    print(s.is_empty())
    print(s)
    print(s.pop())
    print(s.size())
    # print(s.top())
    print(s)
test_array_stack()
```

```python
class Full(Exception):
    pass
class ArrayStack2(Stack):
    def __init__(self, size=0) -> None:
        self.__array = [None] * size
        self.__total_ops = 0
        self.__capacity = size
        self.__n = 0
    def __total_ops_pp(self):
        self.__total_ops+=1
    def __check_is_empty(self):
        if self.is_empty():
            raise Empty("Stack is empty")
    def __check_is_full(self):
        if self.is_full():
            raise Full("Stack is full")
    def __len__(self):
        return self.__n
    def is_empty(self):
        return len(self) == 0
    def is_full(self):
        return self.__n == self.__capacity - 1
    def __n_pp(self):
        self.__n+=1
    def __n_mm(self):
        self.__n-=1
    def capacity(self):
        return self.__capacity
    # def increase_capacity(self, additional_capacity):
    #     self.__capacity += additional_capacity
    def push(self, item):
        self.__check_is_full()
        self.__array[self.__n] = item
        self.__n_pp()
    def pop(self):
        self.__check_is_empty()
        item = self.__array[self.__n - 1]
        self.__array[self.__n - 1] = None
        self.__n_mm()
        return item
    def top(self):
        self.__check_is_empty()
        return self.__array[-1]
    def size(self):
        return len(self)
    def __str__(self):
        return '# Contents: ' + str(self.__array) + '\n# Total Operations: ' + str(self.__total_ops)


def test_array_stack_2():
    s = ArrayStack2(10)
    s.push(1)
    print(s.top())
    print(len(s))
    print(s.size())
    print(s.is_empty())
    print(s)
    print(s.pop())
    print(s.size())
    # print(s.top())
    print(s)
test_array_stack_2()
```

```python
class Parenthesis():
    def __init__(self, left = '(', right = ')'):
        self.left = left
        self.right = right
def matching_parentheses(string, parenthesis=Parenthesis()):
    s = ArrayStack()
    for char in string:
        if char == parenthesis.left:
            print(end=' ')
            s.push(char)
        elif char == parenthesis.right:
            if s.is_empty():
                print()
                return False
            else:
                s.pop()
        else:
            print(char, end='')
    print()
    return s.is_empty()


def test_matching_parentheses(s, parenthesis=Parenthesis()):
    print(matching_parentheses(s, parenthesis))
test_matching_parentheses('(hello(world))')
test_matching_parentheses('((hello(world))')
test_matching_parentheses('(hello(world)))')

test_matching_parentheses('{hello{world}}', Parenthesis('{', '}'))
test_matching_parentheses('ahelloaworldbb', Parenthesis('a', 'b'))
```

```python
def do_parentheses_match(input_string):
    s = []
    balanced = True
    index = 0
    while index < len(input_string) and balanced:
        token = input_string[index]
        if token == "(":
            s.append(token)
        elif token == ")":
            if len(s) == 0:
                balanced = False
            else:
                s.pop()

        index += 1

    return balanced and len(s) == 0
```

```python
def is_matched(expr):
    lefty = '{[('
    righty = '}])'
    s = ArrayStack()
    for c in expr:
        if c in lefty:
            s.push(c)
        elif c in righty:
            if s.is_empty():
                return False
            if righty.index(c) != lefty.index(s.pop()):
                return False
    return s.is_empty()


def test_is_matched(expr):
    print(is_matched(expr))
test_is_matched('(hello(world))')
test_is_matched('{hello{world}}')
test_is_matched('[[[hello]world]')
test_is_matched('[[hello]world]]')
            
```

```python
def is_matched_html(raw):
    s = ArrayStack()
    j = raw.find('<')
    while j != -1:
        k = raw.find('>', j+1)
        if k == -1:
            print('Unmatched <')
            return False
        tag = raw[j+1:k]
        if not tag.startswith('/'):
            s.push(tag.split(' ')[0])
        else:
            if s.is_empty():
                print('Unmatched >')
                return False
            op = s.pop()
            if tag[1:].split(' ')[0] != op:
                print('different tags: '+ tag[1:].split(' ')[0] + ' ' + op)
                return False
        j = raw.find('<', k+1)
    return s.is_empty()


def test_is_matched_html(raw):
    print(is_matched_html(raw))
test_is_matched_html('<html><head><title>hello</title></head><body><h1>hello</h1></body></html>')
with open('Google.html') as f:
    raw = f.read()
    test_is_matched_html(raw)
```
